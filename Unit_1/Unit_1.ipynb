# 6.9 Задание 2. Продолжаем работать с клубами, в которых средняя зарплата совпадает с медианной. Каков максимальный размер средней зарплаты в этой группе клубов? Как называется клуб, где игроки получают такую зарплату?

```python
import pandas as pd
football = pd.read_csv('D:\Data Scientist_Skill Factory/data_sf.csv')
s = (football.groupby(['Club']).Wage.mean() - football.groupby(['Club']).Wage.median() == 0)
club_ls = [] #Создаем пустой список
wage_ls = [] #Создаем пустой список
for i in range (0,649): #Этим циклом мы для каждого клуба проверяем равенство средей и медианной зарплаты и если они равны, то
    if s[s.index[i]] == True:#в пишем название клуба в список club_ls, а среднюю зарплату в список wage_ls. Индексы i совпадают.
        club_ls.append(s.index[i])
        wage_ls.append(football[football.Club == s.index[i]].Wage.mean())
print(max(wage_ls))
i = 0
while wage_ls[i] != max(wage_ls): #Этим циклом мы перебираем список wage_ls пока не находим максимальный элемент, а найдя его 
    i += 1 # по индексу i находим соответствующий ему элемент из списка club_ls 
print(club_ls[i])
```
